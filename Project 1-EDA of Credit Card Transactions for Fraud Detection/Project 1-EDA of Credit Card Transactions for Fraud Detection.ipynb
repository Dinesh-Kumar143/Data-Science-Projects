#### **Question 8: Can we create a bar chart showing the count of fraudulent vs. legitimate transactions?**
**Answer: Yes we can create a bar chart showing the count of fraudulent vs. legitimate transactions but The number of  Fraud cases are very low because of that fraud amount is not visible but, we can create Pie chart for better understand and better visibility as shown below bar chart.**

```python
plt.bar(['Legitimate', 'Fraud'], [total_Legitimate, total_fraud], color=['b', 'r'])
```

```python
plt.figure(facecolor='black')
plt.pie([total_Legitimate, total_fraud], labels=['Legitimate', 'Fraud'], autopct='%1.3f%%', colors=['y', 'r'])
```

#### **Question 10: Can we use a heatmap to visualize the correlation between numerical features?**
**Answer: Yes, we can visualize the correlation between numerical features with heat map where the chart uses color intensity to represent the strength and direction of correlations with digonal showing perfect self-correlation with value 1.**

```python
plt.figure(figsize=(8,6))
plt.imshow(df.corr(), cmap='hot', interpolation='nearest')
```
